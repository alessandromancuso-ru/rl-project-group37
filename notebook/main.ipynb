# 0. Environment

Declare your environment here. Feel free to use add any code cells.

```python
# Import gymnasium and environment definition
env = gym.make('Blackjack-v1', natural=False, sab=False)

def basic_policy(state):
    player_sum, dealer_card, usable_ace = state
    
    if player_sum >= 20:
        return 0   # Stick
    else:
        return 1    # Hit

def print_real_cards(env):
    dealer_cards = env.unwrapped.dealer
    player_cards = env.unwrapped.player
    
    print("Player cards:", player_cards)
    print("Dealer cards:", dealer_cards)
    print("--------------------------------------------------------------")
    
def print_state(state):
    player_sum, dealer_card, usable_ace = state
    print("Player sum: ", player_sum, env.unwrapped.player)
    print("Dealer visible card: ", dealer_card)
    print("Usable ace: ", usable_ace == 1)
    print("--------------------------------------------------------------")
    
def print_result(result):
    print("End game:")
    print("Final Player Cards:", env.unwrapped.player)
    print("Final Dealer Cards:", env.unwrapped.dealer)
    if(result == 0):
        print("Draw")
    elif(result > 0):
        print("Win")
    elif(result < 0):
        print("Lose")

print("Start Game! ")
# Reset the environment to start
observation, info = env.reset()
# Print observation
print_state(observation)

episode_over = False
total_reward = 0

while not episode_over:
    # Take an action based on the initial observation
    action = basic_policy(observation)
    
    print("Action taken: ")
    if(action == 1):
       print("Hit!")
    elif(action == 0):
       print("Stick!")
    
    observation, reward, terminated, truncated, info = env.step(action)
    print_state(observation)
    
    total_reward += reward
    episode_over = terminated or truncated

# print(f"Episode finished! Total reward: {total_reward}")
print_result(total_reward)

env.close()

```
